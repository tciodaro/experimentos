    </div>
    <div style="margin-left:2cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        Swarm intelligence are a sort of metaheuristics algorithms made to solve optimization problems. The main difference from other algorithms is the biological inspiration, mainly in the social behavior of collective agents. Most observations in nature support the theory that multiple agents, with self interests, work collectively without the need of a centralized command, in cooperation to achieve a desired behavior. Notice that not all population-based algorithms are considered Swarm Intelligence, like genetic algorithms.
        </p>
        <p>
        Some mechanisms that are used in almost all swarm algorithms and affect directly the algorithm performance.
        </p>
        <img style="float:right" src="static/concept_blocks.png"/>
        <ul>
            <li>Exploration and Exploitation: local x global</li>
            <li>Diversity: exploratory capacity, affecting the possibility of finding new solutions</li>
            <li>Communication: how social agents interact to each other </li>
            <li>Evaluation/selection: excluding poor solutions, allowing possible non-differentiable cost functions</li>
            <li>Randomness: stochastic search, to avoid local optimals</li>
            <li>Parallelism: neighborhood/communities </li>
        </ul>

        <h1>PSO - Particle Swarm Optimization</h1>
    </div>
    <div style="margin-left:1cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        PSO is a type of algorithms for swarm intelligence that uses the concept of particles in order to find the optimal value of a set of parameters that maximizes/minimizes a specific cost function. The cost function does not need to be differentiable considering the parameters, but it must be able to evaluate the particle according to its position in a search-space. This space consists of the hyper-dimension derived from all possible values for the set of parameters. This article describes the most common PSO algorithm, which can be found at [1]
        </p>
        <p>
        The particles are free to move around the search-space (which can be bounded, for example, when some constraints are considered for the possible solution). The algorithm search capacity is directly affected by the particle dynamics: each particle has an associated speed and memory. The swarm has a social memory, consisting of the best position ever found by the swarm (any particle), while each particle has its own memory (its best position ever). These memories are called social and cognitive memory respectively and, together with the particle previous speed, are used to calculated the next speed vector.
        </p>
        <div style="margin-left:0cm;color:#444444">
            <h2>Mathematical Model</h2>
        </div>
         <div style="margin-left:1cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
            <p>
            Let ${\bf A} \subset \Re^k$ be the search space, in a $k$ dimensional hyperspace, where the particle $\vec{x} = (x_1,...,x_k)$ represents a single point in this space. The algorithm needs an evaluation function, or cost function, that transforms the search space into a one-dimensional space ${\bf Y}$: $$f : {\bf A} \rightarrow {\bf Y} \subseteq \Re,\hspace{1cm} f_i = f(\vec{x}_i) \in {\bf Y}$$
            <br>
            The particle swarm of size $N$ is written as $\textbf{S}(t) = \{ \vec{x}_1(t),...,\vec{x}_N(t) \}$, while its dynamics can be written as the first derivative (speed) of each particle: $\textbf{S'}(t) = \{ \vec{v}_1(t),...,\vec{v}_N(t) \}$. Notice that the particle position and speed are functions of the time $t$, which represents the algorithm iterations. The basic PSO algorithm also requires the definition of the social and cognitive memories:<br>
           $\hspace{3cm}\textbf{P} = \{ \vec{p}_1(t), ..., \vec{p}_N(t) \} \hspace{1cm} \rightarrow \hspace{1cm} \text{cognitive}$
           <br>
           $\hspace{3cm}\vec{p}_i(t) = argmin_{t}f(\vec{x}_i(t))$
           <br><br>
           $\hspace{3cm}\vec{p}_g(t) = argmin_t f(\vec{p}_i(t))\hspace{0.9cm} \rightarrow \hspace{1cm} \text{social}$
            </p>
            <p>
            The swarm dynamics is update at each iteration of the algorithm. The position of each particle is updated considering its speed vector, while the speed vector itself is updated considering the following equation:
            <br><br>
            $$\vec{v}_i(t+1) = w\vec{v}_i(t) + c_1R_1(\vec{p}_i(t) - x_i(t)) + c_2R_2(\vec{p}_g(t)-x_i(t))$$
            <br>
            The first term correspond to the current speed, weighted by $w$, called the inertial weight. This term is responsible for allowing a fine search when the current particle is the global best. Otherwise, this particle would not move anymore (the second and third terms would be zero). Normally, the inertial weight changes in time, in order to allow for global and local searches, affecting the swarm diversity:
            <br><br>
            $$w(t) = W_{up} - (w_{up} - w_{low}) \times \frac{t}{T_{max}}$$
            <br>
            The second term in the speed formula represents the cognitive memory, as it depends on the particle best position ever found. The term $c_1R_1$ adds a randomness search in the direction of $(\vec{p}_i(t) - x_i(t))$. The third term is similar to the second, but it accounts for the social memory. The $R_1$ and $R_2$ terms are random vectors generated from an uniform distribution between 0 and 1 (other random generators can also be used), that multiplies each speed component in the $k$-dimensional space. The $c_1$ and $c_2$ coefficients must be chosen carefully in order to keep the swarm stable.
            </p>
            <p>
            Another variation of the swarm update algorithm uses the constriction coefficient. This formula was suggested after the studies in [2] and adds a coefficient before updating the speed vector:
            <br><br>
            $$\vec{v}_i(t+1) = \chi [\vec{v}_i(t) + c_1R_1(\vec{p}_i(t) - x_i(t)) + c_2R_2(\vec{p}_g(t)-x_i(t))]$$
            <br>
            The standard values for the coefficients are:
            $$\chi = 0.729$$
            $$c_1 = 2.05$$
            $$c_2 = 2.05$$
            
            </p>
        </div>
        <div style="margin-left:1cm;color:#444444">
            <h3>Comments</h3>
        </div>
        <div style="margin-left:1cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
            <p>
            If the search space is bounded, because of constraints in the solution candidades, por example, then there are two common strategies:
            </p>
            <ol>
                <li>Truncate: the solution is truncated and the particle is kept in the limits of the search space </li>
                <li>Bounce-back: the search space limits work as a mirror, bouncing the particle position back to the interior of the allowed search space.
                </li>
            </ol>
            <p>
            The speed is also limited in order assist in the swarm stability. If the speed is greater, or lower, than a specific value, than the new speed value is clipped to this specific value. The strategy suggested is to limit the speed considering a fraction of the search space.
            </p>
        </div>

    <div style="margin-left:2cm;color:#333333">
        <h3>Cost Function</h3>
    </div>
    <div style="margin-left:3cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        The cost function for this test is made of $ngauss$ Gaussian of 2 dimensions, with negative sign, random mean and variance. This creates a search space with several valleys, in order to illustrate a problem with several local optimal solutions. The idea is just to show the dynamics of the PSO algorithm.
        </p>
    </div>
</body></html>

```python
import os
import sys
import matplotlib.pyplot as plt

if f"{os.getcwd()}/../" not in sys.path:
    sys.path.append(f"{os.getcwd()}/../")


from Swarm.Problem_Gaussian import Problem
%matplotlib inline

ngauss = 20
problem = Problem(ngauss, -1)
plt.figure(figsize=(10,6))
problem.plot(colorbar=True)
plt.title('Cost Function', fontsize=18, fontweight='bold');
```

<html>
    <body>

    <div style="margin-left:2cm;color:#333333">
        <h3>Solver</h3>
    </div>
    <div style="margin-left:3cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        The PSO algorithm solver can be started with the parameters given below. It can be seen that the solver was configured considering the constriction coefficient. The swarm was configured with $nswarm = 20$ particles and with a coefficient of speed limit as $k = 1$. The swarm is randomly initialized across the search space and the social and cognitive memories are set.
        </p>
    </div>
</body></html>

```python
from Swarm import PSO
import numpy as np

swarm_dim = problem.get_problem_dim()
swarm_min, swarm_max = problem.get_solution_limits()
psoArgs = {
    'nswarm': 20,
    'swarm_dim': swarm_dim,
    'epochs': 20,
    'chi': 0.729,
    'rate_cog': 2.05,
    'rate_soc': 2.05,
    'K': 1.0,
    'Wup': 1.5,
    'Wlo': 0.5,
    'Wstep': 200,
    'swarm_max': swarm_max,
    'swarm_min': swarm_min,
    'speed_lim': -1
}

mypso = PSO.PSO(psoArgs)
mypso.cost_func = problem.cost_function
seed = 2
mypso.initialize(seed=seed)
# Initial Swarm
plt.figure(figsize=(10,6))
problem.plot(mypso.swarm)
plt.title('Initial Swarm', fontsize=18, fontweight='bold');
```

<html>
    <body>
    <div style="margin-left:3cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        The algorithm's exploitation and exploration capacities are the key points for performance, together with the efficiency of information sharing among the swarm. Those aspets give the stochastic search for the global optimal, leading to an efficient optimization algorithm. The following image shows the possible search space for a given particle, at the first iteration of the algorithm. As it can be seen, the randomness of the search is controlled by the cognitive and social coefficients, later weighted by the constriction coefficient. The green dot shows the best position for the swarm, while the blue dots show the possible next positions for the $i$-th particle. The black dots show the position of the swarm particles, together with its index. In this example, only the social component of the speed is considered, but the next particle position also considers the cognitive and inertial components. 
        </p>
    </div>
</body></html>

```python
plt.figure(figsize=(10,6))
problem.plot(mypso.swarm)
N = 200
S = np.array(mypso.swarm)
V = np.zeros(S.shape)
R1 = np.random.rand(N, mypso.nswarm, mypso.swarm_dim)
R2 = np.random.rand(N, mypso.nswarm, mypso.swarm_dim)
P = np.array(S)
Pcost = mypso._PSO__cost()
gbest = np.argmin(Pcost)
gbest_cost = Pcost[gbest] 
mypso.rate_cog = 0
plt.plot(P[gbest,0],P[gbest,1],'go')
i = 10
for it in range(N):
    V[i] = mypso.chi*(mypso.rate_cog * R1[it][i] *(P[i]    - S[i]) +\
                      mypso.rate_soc * R2[it][i] *(P[gbest]-S[i]))
    nextS = S + V
    plt.plot(nextS[i,0], nextS[i,1],'.b')
plt.axis([mypso.swarm_min, mypso.swarm_max, mypso.swarm_min, mypso.swarm_max])
plt.title('Particle %i Stochastic Search'%(i),  fontsize=18, fontweight='bold');
```

<html>
    <body>
    <div style="margin-left:3cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        The algorithm time performance is mainly affected by the complexity of the cost function. As one of its advantage, the PSO algorthm does not require the cost function to be differentiable, but it must be possible for the algorithm to evaluate the entire search space. 
        <br>
        <br>
        After the algorithm training, where all the iterations are executed, or some other stop criteria is met, the swarm configuration can be analyzed. In this 2D problem, the visualization is straight forward. If the swarm is still spread across the searching space, the algorithm convergence and performance can be questioned as, in general, the swarm should shrink around the optimal value found. Although, it is possible that the swarm passed by the optimal value and position along its evolution. If the parameters are well tuned, this should not happen. The left figure shows the final swarm configuration for the 2D problem. It can be seen that the swarm is gathered around the optimal position.
        <br>
        <br>
        Another interesting behavior to inspect is the swarm best position evolution, pictured in the centered figure. It can be seen, in our example, that the algorithm was able to walk through peaks and valleys in order to find the global optimal.
        <br>
        <br>
        Finally, the right figure shows the final best position found by the algorithm, together with the analytical position of the global optimal for this problem. It can be seen that the two positions are quite close to each other.
        </p>
    </div>
</body></html>

```python
(solution_idx, sol_cost) = mypso.train()
opt = problem.get_optimal()

print('Analytical Optimal: %.2f, at position (%.2f, %.2f)'%(opt[1], opt[0][0], opt[0][1]))
print("Solution found with cost: %.2f, at position (%.2f, %.2f)"%(sol_cost, mypso.swarm_best[-1][0],
                                                                  mypso.swarm_best[-1][1]))

plt.figure(figsize=(18,6))
plt.subplot(1,3,1)
problem.plot(mypso.swarm)
plt.plot(opt[0][0], opt[0][1], 'sr', ms=8)
plt.title('Final Swarm Position', fontsize=18, fontweight='bold')

opt_solution = mypso.swarm_best[-1]
plt.subplot(1,3,2)
problem.plot(mypso.swarm_best, fmt='-ob')
plt.plot(opt[0][0], opt[0][1], 'sr', ms=8)
plt.title('Swarm Optimal Evolution', fontsize=18, fontweight='bold')

plt.subplot(1,3,3)
plt.plot(opt[0][0], opt[0][1], 'sr', ms=8)
problem.plot(mypso.swarm_best[-1], 'sg')
plt.title('Final Optimal Found', fontsize=18, fontweight='bold');
```

    <div style="margin-left:3cm;margin-right:2cm;margin-top:1cm;color:#777777;clear:left">
        <p>
        As usually shown in optimization algorithms, the best cost function value found at each iteration can be plotted to analyze how difficult it was for the algorithm to evolve (left figure). Considering the PSO algorithm, the swarm diversity is a good measure of how well tuned are the swarm exploitation and exploration capabilities. Ideally, the swarm should be able to exploit the entire search space in order to spot possible local optima positions. Later, as the swarm diversity reduces, as the algorithm iterates, the swarm explores these regions in order to find the global optimum. The swarm inertial weight is suitable for controlling the exploitation / exploration characteristics, as its value reduces in time. The swarm diversity, though, should reduce in time if the swarm is evolving correctly. Other mechanisms to bump up the swarm diversity randomly in time could be used to force the swarm to explore other regions, specially when the algorithm convergence seems to be stalled in a local optimal.
        </p>
